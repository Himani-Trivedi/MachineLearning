2.) Write Python code to multiply the above two matrices. Solve using list of lists and then use numpy. Compare the timing of both solutions. Which one is faster? Why?

Numpy is faster then nested list as numpy is specifically created for mathematical compution and stores element in contigous memory location unlike lists which requires extra overhead for accessing address of next element.

```python
import numpy as np
import datetime as dt


arr=np.array([[2,34,56],[1,5,67],[23,45,6]])
arr2=np.array([[45,7],[1,5],[45,6]])

start=dt.datetime.now()
np.dot(arr,arr2)
end=dt.datetime.now()
print("Time with Numpy" , (end-start).total_seconds())

n=3
q=3
m=2

A=[[2,34,56],[1,5,67],[23,45,6]] 
B=[[45,7],[1,5],[45,6]] 
Mul=[]

start=dt.datetime.now()
for i in range(n):
    temp=[]
    for j in range(m):
        elemSum=0
        for k in range(q):
            elemSum+=A[i][k] * B[k][j]
        temp.append(elemSum)
    Mul.append(temp)
end=dt.datetime.now()
print("Time without Numpy" , (end-start).total_seconds())
print(Mul)
```

What is the gradient of the following function with respect to x and y?
f(x,y)=x^2y + y^3sin(x)

gradient of f(x,y) is (2xy + y cos(x), x^2 + 3y^2 sin(x))


Use JAX to confirm the gradient evaluated by your method matches the analytical solution corresponding to a few random values of x and y

```python
%pip install jax
```

```python
import jax
import jax.numpy as jnp

def f(x, y):
    return x**2 * y + y**3 * jnp.sin(x)

grad_f = jax.grad(f, argnums=(0, 1))

x_val = 2.0
y_val = 3.0

analytical_gradient = grad_f(x_val, y_val)
print("Analytical Gradient:", analytical_gradient)
numerical_gradient = jax.grad(lambda x, y: f(x, y))(x_val, y_val)
print("Numerical Gradient:", numerical_gradient)
```

```python
import sympy as sp

x, y = sp.symbols('x y')
f = x**2 * y + y**3 * sp.sin(x)

grad_f_x = sp.diff(f, x)
grad_f_y = sp.diff(f, y)


print("Symbolic Gradient:")
print("Partial derivative with respect to x:", grad_f_x)
print("Partial derivative with respect to y:", grad_f_y)
```

Store the same information using Python classes. We have an overall database which is a list of year objects. Each year contains a list of branches. Each branch contains a list of students. Each student has some properties like name, roll number and has marks in some subjects.

```python
class Student:
    def __init__(self, name, roll_number, marks):
        self.name = name
        self.roll_number = roll_number
        self.marks = marks

    def __str__(self):
        return f"Student(Name: {self.name}, Roll Number: {self.roll_number}, Marks: {self.marks})"


class Branch:
    def __init__(self, name, students):
        self.name = name
        self.students = students

    def __str__(self):
        return f"Branch(Name: {self.name}, Students: {len(self.students)})"


class Year:
    def __init__(self, year, branches):
        self.year = year
        self.branches = branches

    def __str__(self):
        return f"Year(Year: {self.year}, Branches: {len(self.branches)})"


class Database:
    def __init__(self, years):
        self.years = years

    def __str__(self):
        return f"Database(Years: {len(self.years)})"

student1 = Student("Alice", "A101", {"Math": 90, "Physics": 85, "Chemistry": 92})
student2 = Student("Bob", "B102", {"Math": 88, "Physics": 89, "Chemistry": 78})

branch1 = Branch("Computer Science", [student1, student2])
branch2 = Branch("Electrical Engineering", [Student("Charlie", "C103", {"Math": 75, "Physics": 80, "Chemistry": 85})])

year1 = Year("2022", [branch1])
year2 = Year("2023", [branch1, branch2])

database = Database([year1, year2])
print(database.years[0].branches[0].students[0])
```

Add a new column to the dataframe called “f” which is the sum of the columns “a”, “b”, “c”, “d”, “e”. Create another column called “g”. The value in the column “g” should be “LT8” if the value in the column “f” is less than 8 and “GT8” otherwise. Find the number of rows in the dataframe where the value in the column “g” is “LT8”. Find the standard deviation of the column “f” for the rows where the value in the column “g” is “LT8” and “GT8” respectively.

```python

df["f"] = df[["a", "b", "c", "d", "e"]].sum(axis=1)
df["g"] = pd.cut(df["f"], bins=[0, 8, 100], labels=["LT8", "GT8"])

num_lt8 = df[df["g"] == "LT8"].shape[0]

sd_lt8 = df[df["g"] == "LT8"]["f"].std()
sd_gt8 = df[df["g"] == "GT8"]["f"].std()

# Print results
print(f"Number of rows with 'g' as 'LT8': {num_lt8}")
print(f"Standard deviation of 'f' for 'LT8': {sd_lt8}")
print(f"Standard deviation of 'f' for 'GT8': {sd_gt8}")

```
